```python
import nltk
from nltk.corpus import gutenberg
fieldids_gutenberg = nltk.corpus.gutenberg.fileids()

def remove_punctuation(corpus):
        cleaned_corpus=[]

        punctuations = ['!','(',')','-','[',']','{',';',':',"'",'\\','<','>','.','/','?','~','&',"''",',','--','``','"']
        for sent in corpus :
            sent1=[]
            for word in sent:
                if word not in punctuations:
                    sent1.append(word)
            cleaned_corpus.append(sent1)  
        return cleaned_corpus
    
corpus_gutenberg = remove_punctuation(list(gutenberg.sents(fieldids_gutenberg)))
```

```python
#flattening a list
gutenberg_tokens = [item for sublist in corpus_gutenberg for item in sublist]
print(gutenberg_tokens[:200])



print('Total Tokens: %d' % len(gutenberg_tokens))
print('Unique Tokens: %d' % len(set(gutenberg_tokens)))
```

```python
# organize into sequences of tokens
length = 50 + 1
sequences = list()
for i in range(length, len(gutenberg_tokens)):
    # select sequence of tokens
    seq = gutenberg_tokens[i-length:i]
    # convert into a line
    line = ' '.join(seq)
    # store
    sequences.append(line)
print('first Sequence \n',  sequences[0])
print('second Sequence \n',  sequences[1])

print('Total Sequences: %d' % len(sequences))
```

```python
# save tokens to file, one dialog per line
def save_doc(lines, filename):
	data = '\n'.join(lines)
	file = open(filename, 'w')
	file.write(data)
	file.close()



# save sequences to file
out_filename = 'gutenberg_sequences.txt'
save_doc(sequences, out_filename)
```

```python
# load doc into memory
def load_doc(filename):
	# open the file as read only
	file = open(filename, 'r')
	# read all text
	text = file.read()
	# close the file
	file.close()
	return text

# load
in_filename = 'gutenberg_sequences.txt'
doc = load_doc(in_filename)
lines = doc.split('\n')
```

```python
print(set([len(i) for i in lines]))
```

```python
from keras.preprocessing.text import Tokenizer
# integer encode sequences of words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
sequences = tokenizer.texts_to_sequences(lines)
print(type(sequences))
```
